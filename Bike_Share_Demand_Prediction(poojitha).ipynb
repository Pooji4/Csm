```python
#Data Preparation

import pandas as pd

# Assuming 'data' is your DataFrame
data = pd.read_csv('bike_share.csv')

# Check column names
print(data.columns)

# Drop columns if they exist
columns_to_drop = ["instant", "dteday", "casual", "registered"]
columns_to_drop = [col for col in columns_to_drop if col in data.columns]  # Ensure columns exist
data.drop(columns=columns_to_drop, axis=1, inplace=True)

data.head()
```

```python
# dummy variables
season_type = pd.get_dummies(data['season'], drop_first = True)
season_type.rename(columns={2:"season_summer", 3:"season_fall", 4:"season_winter"},inplace=True)
season_type.head()
```

```python
# Create dummy variables for 'weathersit' column
weather_type = pd.get_dummies(data['weathersit'], prefix='weather', drop_first=True)

# Rename columns
weather_type.rename(columns={2: "weather_mist_cloud", 3: "weather_light_snow_rain"}, inplace=True)

weather_type.head()
```

```python
# Create dummy variables for 'season' and 'weathersit' columns
season_type = pd.get_dummies(data['season'], prefix='season', drop_first=True)
season_type.rename(columns={2: "season_summer", 3: "season_fall", 4: "season_winter"}, inplace=True)

weather_type = pd.get_dummies(data['weathersit'], prefix='weather', drop_first=True)
weather_type.rename(columns={2: "weather_mist_cloud", 3: "weather_light_snow_rain"}, inplace=True)

# Concatenate dummy variables to the main DataFrame
data = pd.concat([data, season_type, weather_type], axis=1)

# Drop original 'season' and 'weathersit' columns
data.drop(columns=["season", "weathersit"], axis=1, inplace=True)

# Analyze DataFrame after dropping columns
data.info()
```

```python
# Check column names
print(data.columns)

# Creating year_quarter derived columns from month columns (if 'mnth' exists)
if 'mnth' in data.columns:
    data["Quarter_JanFebMar"] = data["mnth"].apply(lambda x: 1 if x <= 3 else 0)
    data["Quarter_AprMayJun"] = data["mnth"].apply(lambda x: 1 if 4 <= x <= 6 else 0)
    data["Quarter_JulAugSep"] = data["mnth"].apply(lambda x: 1 if 7 <= x <= 9 else 0)
    # Drop column 'mnth' since we have already created quarters
    data.drop(columns=["mnth"], axis=1, inplace=True)

# Creating 'weekend' column (if 'weekday' exists)
if 'weekday' in data.columns:
    data["weekend"] = data["weekday"].apply(lambda x: 1 if x == 0 or x == 6 else 0)
    # Drop column 'weekday'
    data.drop(columns=["weekday"], axis=1, inplace=True)

# Drop column 'workingday' (if it exists)
data.drop(columns=["workingday"], axis=1, inplace=True, errors='ignore')

data.head()
```

```python
# Building the Linear Model

import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.feature_selection import RFE

# Assuming X_train and y_train are already defined

# Initialize the linear regression model
lm = LinearRegression()
```

```python
# Initialize RFE with the linear regression model and the desired number of features to select
rfe = RFE(estimator=lm, n_features_to_select=12)

# Fit the RFE model on the training data
rfe.fit(X_train, y_train)

# Extract the column names of the selected features
columns_rfe = X_train.columns[rfe.support_]
```

```python
# Create a new dataframe with the selected features
X_train_lr5 = X_train[columns_rfe]
```

```python
from sklearn.linear_model import LinearRegression

# Assuming X_train_lr5 and y_train are already defined

# Initialize the linear regression model
lr5 = LinearRegression()

# Train the final model lr5 on the selected features
lr5.fit(X_train_lr5, y_train)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Use the final model lr5 to predict y_train values (assuming lr5 and X_train_lr5 are defined)
y_train_cnt = lr5.predict(X_train_lr5)

fig = plt.figure(figsize=(8, 6))
sns.histplot((y_train - y_train_cnt), 
             bins=20,         # Number of bins for the histogram
             kde=True,       # Whether to plot a kernel density estimate
             color='skyblue',# Color of the bars
             edgecolor='black',  # Color of the edges
             linewidth=1.2)  # Width of the bar edges

fig.suptitle('Error Terms', fontsize=20)
plt.xlabel('Errors', fontsize=18)
plt.ylabel('Frequency', fontsize=18)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```
